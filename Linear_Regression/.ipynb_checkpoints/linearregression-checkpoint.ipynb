```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Salary_Data.csv')
df
```

```python
def fwbx(w,b,x):
    return np.dot(w,x)+b
def cost(x,y,w,b):
    m=len(x)
    sum1=0
    for i in range(m):
        sum1+=(fwbx(w,b,x[i])-y[i])**2
    return (1/(2*m))*sum1
def gradient_descent(x, y, w, b, alpha=0.01, iterations=1000):
    m = len(x)  # number of samples
    n = x.shape[1] if len(x.shape) > 1 else 1  # number of features
    costs = []
    
    for i in range(iterations):
        # Forward pass (vectorized)
        y_pred = fwbx(w, b, x)
        cost = cost(x, y, w, b)
        costs.append(cost)
        
        # Gradients (vectorized)
        dw = (1/m) * np.dot(x.T, (y_pred - y))  # Shape matches w
        db = (1/m) * np.sum(y_pred - y)
        
        # Update parameters
        if isinstance(w, (int, float)):  # Univariate case
            w -= alpha * dw
        else:  # Multivariate case
            w -= alpha * dw.flatten()
        b -= alpha * db
        
    return w, b, costs

```

```python
X=df['YearsExperience']
Y=df['Salary']
```

```python
w=0
b=0
w,b,costs=gradient_descent(X,Y,w,b)
```

```python
plt.plot(range(len(costs)),costs,marker='x')
plt.xlabel('Iteration')
plt.ylabel('Cost Value')
plt.title('Linear Regression')
plt.show()
```
