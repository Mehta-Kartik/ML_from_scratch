```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Salary_Data.csv')
df
```

```python
def fwbx(w,b,x):
    return np.dot(x,w)+b
def compute_cost(x,y,w,b):
    m=len(x)
    yprd=fwbx(w,b,x)
    sum1=np.sum((yprd-y)**2)
    return (1/(2*m))*sum1
def gradient_descent(x, y, w, b, alpha=0.01, iterations=1000):
    m = len(x)  # number of samples
    n = x.shape[1] if len(x.shape) > 1 else 1  # number of features
    costs = []
    
    for i in range(iterations):
        # Forward pass (vectorized)
        y_pred = fwbx(w, b, x)
        cost = compute_cost(x, y, w, b)
        costs.append(cost)
        
        # Gradients (vectorized)
        dw = (1/m) * np.dot(x.T, (y_pred - y))  # Shape matches w
        db = (1/m) * np.sum(y_pred - y)
        
        # Update parameters
        if isinstance(w, (int, float)):  # Univariate case
            w -= alpha * dw
        else:  # Multivariate case
            w -= alpha * dw.flatten()
        b -= alpha * db
        
    return w, b, costs

```

```python
def mean_squared_error_manual(y_true, y_pred):
    return np.mean((y_true - y_pred)**2)

def root_mean_squared_error_manual(y_true, y_pred):
    return np.sqrt(mean_squared_error_manual(y_true, y_pred))

def mean_absolute_error_manual(y_true, y_pred):
    return np.mean(np.abs(y_true - y_pred))

def r2_score_manual(y_true, y_pred):
    ss_res = np.sum((y_true - y_pred)**2)
    ss_tot = np.sum((y_true - np.mean(y_true))**2)
    return 1 - (ss_res / ss_tot)

def performance_metrics_manual(y_true, y_pred):
    """100% manual regression metrics"""
    mse = mean_squared_error_manual(y_true, y_pred)
    rmse = root_mean_squared_error_manual(y_true, y_pred)
    mae = mean_absolute_error_manual(y_true, y_pred)
    r2 = r2_score_manual(y_true, y_pred)
    
    return {
        'MSE': mse,
        'RMSE': rmse,
        'MAE': mae,
        'R²': r2
    }
```

```python
# Your trained model predictions
y_pred = fwbx(w, b, X)

# Manual metrics
metrics = performance_metrics_manual(Y, y_pred)
for key, value in metrics.items():
    print(f"{key}: {value:.2f}")
```

```python
def ridge_cost(x,y,w,b,alpha):
    firstpart = compute_cost(x, y, w, b)     # your (1/(2m))*sum(...)
    secondpart = alpha * np.sum(w**2)        # L2 on weights only
    return firstpart + secondpart
def lasso_cost(x,y,w,b,alpha):
    firstpart=compute_cost(x,y,w,b)
    secondpart= alpha*np.sum(np.abs(w))
    return firstpart+secondpart
```

```python
def compute_gradient(x, y, w, b):
    m = len(x)
    y_pred = fwbx(w, b, x)
    dw = (1/m) * np.dot(x.T, (y_pred - y))   # shape like w
    db = (1/m) * np.sum(y_pred - y)
    return dw, db

def ridge_gradient_descent(x,y,w,b,alpha):
    dw, db = compute_gradient(x, y, w, b)    # unregularized grads
    dw = dw + 2 * alpha * w                  # ∂/∂w (alpha * sum w^2)
    return dw, db
alpha = 0.01
learning_rate = 0.005
num_iters = 300
w = np.zeros_like(w)     # or scalar 0 for 1D
b = 0.0
for i in range(num_iters):
    djdw,djdb=ridge_gradient_descent(X,Y,w,b,0.04)
    w=w-learning_rate*djdw
    b=b-learning_rate*djdb
    if i%100==0:
        cost=ridge_cost(X,Y,w,b,alpha)
        print(i,cost)
```

```python
y_pred_ridge = fwbx(w, b, X)
metrics_ridge = performance_metrics_manual(Y, y_pred_ridge)
print(metrics_ridge)
```

```python
def lasso_gradient(x, y, w, b, alpha):
    dw, db = compute_gradient(x, y, w, b)
    dw = dw + alpha * np.sign(w)     # L1 subgradient
    return dw, db

# alpha = 0.04
# learning_rate = 0.08
num_iters = 300
w = np.zeros_like(w)
b = 0.0

for i in range(num_iters):
    dw, db = lasso_gradient(X, Y, w, b, alpha)
    w = w - learning_rate * dw
    b = b - learning_rate * db

    if i % 100 == 0:
        cost = lasso_cost(X, Y, w, b, alpha)
        print(i, cost)

y_pred_lasso = fwbx(w, b, X)
metrics_lasso = performance_metrics_manual(Y, y_pred_lasso)
print(metrics_lasso)
```
